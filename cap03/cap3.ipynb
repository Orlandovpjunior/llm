```python

import torch

inputs = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)
```

```python
query = inputs[1]
query
```

```python
inputs.shape
```

```python
attn_scores = torch.empty(inputs.shape[0])
attn_scores
```

```python
for i, x_i in enumerate(inputs):
    attn_scores[i] = torch.dot(x_i, query)
print(attn_scores)
```

```python
attn_weights_temp = attn_scores / attn_scores.sum()
print("Attention weights:", attn_weights_temp)
print("Sum:", attn_weights_temp.sum())
```

```python
attn_scores.shape
```

```python
def softmax_naive(x):
    return torch.exp(x)/ torch.exp(x).sum(dim=0)
attn_weights_naive = softmax_naive(attn_scores)
print("Attentition weights:", attn_weights_naive)
print("Sum:" ,attn_weights_naive.sum())
```

```python
attn_weights = torch.softmax(attn_scores, dim=0)
print("Attention weights:", attn_weights)
print("Sum:",attn_weights.sum())
```

```python
query = inputs[1]
context_vec = torch.zeros(query.shape)
for i, x_i in enumerate(inputs):
    context_vec += attn_weights[i] * x_i
print(context_vec)
```

```python
attn_scores_2 = torch.empty(6,6)
for i, x_i in enumerate(inputs):
    for j, x_j in enumerate(inputs):
        attn_scores_2[i,j] = torch.dot(x_i,x_j)
print(attn_scores_2)
```
